```python
#Import basic packages
import os
import numpy as np
import pandas as pd
import csv
import regex as re
import datetime as dt 
```

```python
#import outcome file with org_uuid
main_df = pd.read_csv(R"d:/msc-project/data/final/outcome_final.csv",encoding='utf-8',
                         parse_dates= ['founded_on','seed_date','series_a_date','series_b_date','series_c_date'])
main_processing_df = main_df[['org_uuid','name','outcome']]
```

```python
#import main organisation csv
org_ori_df = pd.read_csv(R"d:/msc-project/data/pre-processed/organisations_preprocessed.csv",encoding='utf-8',
                         parse_dates= ['founded_on','went_public_on','acquired_on','closed_on'])
```

```python
org_ori_df.columns
```

```python
org_processing_df = org_ori_df[['org_uuid','facebook_url','linkedin_url','twitter_url']]
```

```python
org_processing_df.head(10)
```

```python
#fill NA value with 0 and non NA with 1
org_processing_df['facebook_url'][org_processing_df['facebook_url'].notnull()] = 1
org_processing_df['linkedin_url'][org_processing_df['linkedin_url'].notnull()] = 1
org_processing_df['twitter_url'][org_processing_df['twitter_url'].notnull()] = 1
org_processing_df.fillna(0,inplace=True)
```

```python
org_processing_df.head(10)
```

```python
main_processing_df_v1 = pd.merge(main_processing_df,
                              org_processing_df,
                              on='org_uuid',
                              how='left')
```

```python
main_processing_df_v1.shape
```

```python
main_processing_df_v1.head(10)
```

```python
org_processing_df = org_ori_df[['org_uuid','country_code']]

#create column for one-hot-encode for country
column_name = list()
for country_code in list(pd.get_dummies(org_processing_df.country_code).columns):
    column_name.append(str('is_' + country_code.lower()))
    
country_one_hot = pd.get_dummies(org_processing_df.country_code, columns=column_name)
new_column_dict = dict(zip(country_one_hot.columns,column_name))
country_one_hot.rename(columns = new_column_dict,inplace=True)
```

```python
org_processing_df = pd.concat([org_processing_df['org_uuid'],country_one_hot],
                             axis=1)
org_processing_df.head(3)
```

```python
main_processing_df_v2 = pd.merge(main_processing_df_v1,
                                org_processing_df,
                                on='org_uuid',
                                how='left')
main_processing_df_v2.shape
```

```python
main_processing_df_v2.head(3)
```

```python
#import event appearance file csv
event_appear_df = pd.read_csv(R"d:/msc-project/data/pre-processed/event_appearances_preprocessed.csv",encoding='utf-8')
```

```python
event_appear_df.head(3)
```

```python
top1000_events_uuid = event_appear_df['event_uuid'].value_counts().nlargest(1000)
```

```python
top1000_events_uuid
```

```python
#Create df with top 1000 events (more than 100 occurences)
df=event_appear_df
top_1000_events_df = df[df.groupby('event_uuid')['event_uuid'].transform('size') >= 100]
```

```python
top_1000_events_df.head(3)
```

```python
top1000_attendees = list (set(top_1000_events_df['participant_uuid']))
len(top1000_attendees)
```

```python
is_attend_top1000_event = list()

for org_uuid in list (main_processing_df_v2['org_uuid']):
    if org_uuid in top1000_attendees:
        is_attend_top1000_event.append(1)
    else:
        is_attend_top1000_event.append(0)

main_processing_df_v2['is_attend_top1000_event'] = is_attend_top1000_event
```
