```python
degrees = pd.read_csv("data/degrees.csv")
```

## deal with people data

```python
people_descriptions = people_descriptions[["uuid","name","description"]]
```

```python
people = people[["uuid","name","gender","featured_job_organization_uuid","featured_job_organization_name","featured_job_title"]]
```

```python
people_all = pd.merge(people, people_descriptions, how='left', on=['uuid', 'name'])
```

```python
people_all.columns = ['p_uuid', 'p_name', 'gender', 'uuid',
       'name', 'featured_job_title', 'p_description']
```

```python
people_all
```

## deal with organization data

```python
organizations = organizations[["uuid","name","rank","roles","status","short_description","category_list","category_groups_list","num_funding_rounds","employee_count"]]
```

```python
organization_descriptions = organization_descriptions[["uuid","name","description"]]
```

```python
organizatio_all = pd.merge(organizations, organization_descriptions, how='left', on=['uuid', 'name'])
```

```python
organizatio_all.head()
```

## deal with degree data

```python
degrees = degrees[["uuid","name","person_uuid","person_name","institution_uuid","institution_name","degree_type","subject"]]
```

```python
degrees.columns = ["d_uuid","d_name","p_uuid","p_name","ins_uuid","ins_name","degree_type","subject"]
```

```python
people_all_plus_degree = pd.merge(people_all, degrees, how='left', on=['p_uuid', 'p_name'])
```

```python
people_all_plus_degree.head(5)
```

```python
final_data = pd.merge(organizatio_all, people_all_plus_degree, how='left', on=['uuid', 'name'])
```
