```python
import requests
import pandas as pd
import time
from datetime import datetime
```

```python
corepo_orgs = pd.read_csv('../evaluation/corepo_lookup_preprocessed.csv', index_col=0)
cb_data = pd.read_csv('../evaluation/selected_organizations_crunchbase.csv', index_col=0)
funding_data = pd.read_csv('../../../../data/funding_rounds.csv', index_col=0)

corepo_orgs['last_funding_type'] = None
```

```python
corepo_orgs.drop_duplicates('domain')
```

```python
funding_data['investment_type'].unique()
```

```python
# label each company as prospective startup or not
# startup defined as last funding type being pre-seed, seed, angel, series A
```

```python
for i in range(0, len(corepo_orgs)):
    print(i, end=' ')
    scraped_org = corepo_orgs.iloc[i]
    url = str(scraped_org['domain'])
    if url == 'nan':
        continue
#     print(scraped_org)
#     print(url)
    url = url.split("://")[1]
    url1 = "://" + url
    url2 = "." + url
    #     print(url1, url2)
    # to avoid partial matches, make sure match url has www.url or http(s)://url
    temp = cb_data[cb_data['homepage_url'].str.contains(url1, na=False, regex=False) | cb_data['homepage_url'].str.contains(url2, na=False, regex=False)]
#     temp = cb_data[cb_data['name']=='Carbyne']
    if temp.empty:
        pass
    else:
        temp = temp.iloc[0]
        # corepo_orgs['last_funding_type'].iloc[i] = temp['cb_url']
        org_uuid = temp['uuid']
        temp2 = funding_data[funding_data['org_uuid'] == org_uuid]
        if temp2.empty:
            pass
        else:
            temp2['announced_on'] = pd.to_datetime(temp2['announced_on'])#, format='%m/%d/%Y')
            temp2 = temp2.sort_values(by=['announced_on'], ascending=False)
            value = temp2.iloc[0]['investment_type']
            print(value)
            corepo_orgs['last_funding_type'].iloc[i] = value
        print('')
```
