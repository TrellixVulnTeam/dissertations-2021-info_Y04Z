```python
import pandas as pd
```

```python
# load data 
corepo_orgs = pd.read_csv('corepo_lookup_preprocessed.csv', index_col=0)
corepo_orgs = corepo_orgs.drop(columns = ['search_keyword', 'article_id'])
corepo_orgs = corepo_orgs.drop_duplicates()
corepo_orgs
```

```python
corepo_orgs.drop_duplicates('domain')
```

```python
# load gold standard
cb_data = pd.read_csv('organizations_trimmed_preprocessed.csv', index_col=0)
cb_data
```

```python
corepo_orgs['accuracy'] = 0
```

```python
no_match_count = 566
for i in range(1839, len(corepo_orgs)):
    print(i, end=' ')
    scraped_org = corepo_orgs.iloc[i]
    url = str(scraped_org['domain'])
    if url == 'nan':
        no_match_count += 1
        continue
#     print(scraped_org)
#     print(url)
    url = url.split("://")[1]
    url1 = "://" + url
    url2 = "." + url
    #     print(url1, url2)
    # to avoid partial matches, make sure match url has www.url or http(s)://url
    temp = cb_data[cb_data['homepage_url'].str.contains(url1, na=False, regex=False) | cb_data['homepage_url'].str.contains(url2, na=False, regex=False)]
    if temp.empty:
        no_match_count += 1
    else:
        accuracies = []
        for index, row in temp.iterrows():
            accuracy = 0
            non_empty_columns = 0
            # 1 compare company name
            if str(row.name) != 'nan':
                non_empty_columns += 1
#                 print(row['name'], scraped_org['name'])
                if str(row['name']).lower() in str(scraped_org['name']).lower() or str(scraped_org['name']).lower() in str(row['name']).lower():
                    accuracy += 1
            
            # 2 compare employee count
            if str(row.employee_count)!='unknown':
                non_empty_columns += 1
                if str(row.employee_count) == str(scraped_org.employee_count):
                    accuracy +=1
            # 3 compare funding with accepted error of 10%
            if str(row['total_funding_usd'])!='nan':
                non_empty_columns += 1
                actual = float(row['total_funding_usd'])
                scraped = scraped_org['funding_numeric']
                if scraped != 'unknown':
                    error = abs(actual - float(scraped))/actual
                    if error <= 0.1:
                        # if error is <10%
                        accuracy += 1
            
            # 4 compare country
            if str(row['country_code']) != 'nan':
                non_empty_columns += 1
                actual = row['country_code']
                scraped = scraped_org['country_code']
                if actual == scraped:
                    accuracy += 1
            # 5 compare city
            if str(row['city']) != 'nan':
                non_empty_columns += 1
                actual = row['city']
                scraped = str(scraped_org['city'])
                if actual.lower().strip() == scraped.lower().strip():
                    accuracy += 1
            
            # 6 compare founding year
            if str(row['founded_on_year']) != 'nan':
                non_empty_columns += 1
                actual = int(row['founded_on_year'])
                scraped = scraped_org['founding_year']
                if str(scraped) != 'nan' and int(scraped) == actual:
                    accuracy += 1
                    
            # 7 compare linkedin URL using identifier
            if str(row['linkedin_identifier']) != 'nan':
                non_empty_columns += 1
                actual = row['linkedin_identifier']
                scraped = str(scraped_org['linkedin_identifier'])
                if actual.lower().strip() == scraped.lower().strip():
                    accuracy += 1
                
                    
            accuracies.append(float(accuracy/non_empty_columns))
#             print((accuracy,non_empty_columns))
        print(max(accuracies), end = ' ')
        # store accuracy for the best match
        corepo_orgs['accuracy'].iloc[i] = max(accuracies)
    print('')
    
```
