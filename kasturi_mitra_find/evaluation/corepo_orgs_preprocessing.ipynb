```python
import pandas as pd
import pycountry
import numpy as np
```

```python
corepo_orgs = pd.read_csv('../techcrunch_scrape/corepo_lookup_structured.csv', index_col=0)
corepo_orgs
```

```python
# remove any duplicates
corepo_orgs = corepo_orgs.drop_duplicates()
corepo_orgs
```

```python
cb_data = pd.read_csv('organizations_trimmed.csv', index_col=0)
cb_data
```

```python
print(corepo_orgs['employee_count'].unique())
print(cb_data['employee_count'].unique())
```

```python
# merge employee_count categories in scraped data
corepo_orgs['employee_count'] = corepo_orgs['employee_count'].str.replace(",", "")
corepo_orgs['employee_count'] = corepo_orgs['employee_count'].str.replace(" - ", "-")
corepo_orgs['employee_count'] = corepo_orgs['employee_count'].str.replace(" or more", "+")
corepo_orgs['employee_count'] = corepo_orgs['employee_count'].str.replace('51-200', '51-500' ).replace('201-500', '51-500')
corepo_orgs['employee_count'] = corepo_orgs['employee_count'].replace(np.nan, 'unknown')
print(corepo_orgs['employee_count'].unique())
```

```python
# merge employee_count categories in crunchbase data
cb_data['employee_count'] = cb_data['employee_count'].str.replace(r'^1-10', '1-50', regex=True).replace('11-50', '1-50')
# cb_data['employee_count'] = cb_data['employee_count'].str.replace('5001-50000', '5001-10000').replace('5001-50000', '5001-10000').replace('5001-50000', '5001-10000')
cb_data['employee_count'] = cb_data['employee_count'].str.replace('51-100', '51-500' ).replace('101-250', '51-500').replace('251-500', '51-500')
print(cb_data['employee_count'].unique())
```

```python
# create column with country codes in scraped data
countries = {}
for country in pycountry.countries:
    countries[country.name] = country.alpha_3
corepo_orgs['country_code'] = corepo_orgs['country'].str.lower().replace('united states of america', 'united states').str.title()
corepo_orgs['country_code'] = [countries.get(country, 'unknown') for country in corepo_orgs['country_code']]
corepo_orgs
```

```python
# convert funding in scraped to numbers
funding_numeric = []
for i in range(0, len(corepo_orgs)):
    funding = corepo_orgs.iloc[i]['funding']
    if str(funding) == 'nan':
        funding_numeric.append('unknown')
    else:
        funding = str(funding).strip().split(' ')
        if len(funding)>1 and funding[1].lower() == 'million':
            funding = float(funding[0])*1e6
        elif len(funding)>1 and funding[1].lower() == 'billion':
            funding = float(funding[0])*1e9
        else:
            funding = funding[0]
        funding_numeric.append(funding)
# print(funding_numeric[11], corepo_orgs.iloc[11]['funding'])
        
corepo_orgs['funding_numeric'] = funding_numeric
corepo_orgs
```

```python
# take only the relevant part of linkedin URL in corepo_orgs
corepo_orgs['linkedin_identifier'] = corepo_orgs['linkedin'].str.split('company/', expand=True)[1].str.strip('/')
```

```python
# take only the relevant part of linkedin URL in crunchbase data
cb_data['linkedin_identifier'] = cb_data['linkedin_url'].str.split('company/', expand=True)[1].str.strip('/')
```
